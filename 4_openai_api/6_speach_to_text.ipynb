```python
import os
from google.colab import userdata

# 環境変数の準備 (左端の鍵アイコンでOPENAI_API_KEYを設定)
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
```

```python
from openai import OpenAI

# クライアントの準備
client = OpenAI()
```

# 音声認識の実行

```python
# 音声ファイルの読み込み (音声ファイルをカレントフォルダに配置)
audio_file= open("audio.wav", "rb")
```

```python
# 音声合成の実行
response = client.audio.transcriptions.create(
    model="gpt-4o-mini-transcribe",  # モデルID
    file=audio_file,  # 音声ファイル
)
print(response.text)
```

# プロンプトによる品質向上

```python
# 音声合成の実行
response = client.audio.transcriptions.create(
    model="gpt-4o-mini-transcribe",  # モデルID
    file=audio_file,  # 音声ファイル
    prompt="次の会話は、OpenAI、GPT-5、そしてAIの将来に関する最近の動向についての講義です。",  # プロンプト
)
print(response.text)
```
