```python
import os
from google.colab import userdata

# 환경 변수 준비 (왼쪽 끝 키 아이콘으로 OPENAI_API_KEY 설정)
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
```

```python
from openai import OpenAI

# 클라이언트 준비
client = OpenAI()
```

# 음성 인식 실행

```python
# 음성 파일 읽기 (음성 파일을 현재 폴더에 배치)
audio_file= open("audio.wav", "rb")
```

```python
# 음성 인식 실행
response = client.audio.transcriptions.create(
    model="gpt-4o-mini-transcribe",  # 모델 ID
    file=audio_file,  # 음성 파일
)
print(response.text)
```

# 프롬프트를 통한 품질 향상

```python
# 음성 인식 실행
response = client.audio.transcriptions.create(
    model="gpt-4o-mini-transcribe",  # 모델 ID
    file=audio_file,  # 음성 파일
    prompt="다음 대화는 OpenAI, GPT-5, 그리고 AI의 미래에 관한 최근 동향에 대한 강의입니다.",  # 프롬프트
)
print(response.text)
```
