```python
import os
from google.colab import userdata

# 환경 변수 준비 (왼쪽 끝 키 아이콘으로 OPENAI_API_KEY 설정)
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
```

```python
from openai import OpenAI

# 클라이언트 준비
client = OpenAI()
```

# 텍스트로부터 이미지 생성

```python
import base64

# 이미지 생성
result = client.images.generate(
    model="gpt-image-1",  # 모델 ID
    prompt="귀여운 고양이 귀 메이드가 미소를 지으며 손을 흔들고 있다"  # 프롬프트
)

# 이미지 저장
image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)
with open("sample.png", "wb") as f:
    f.write(image_bytes)
```

# 이미지 출력 사용자화

```python
import base64

# 이미지 생성
result = client.images.generate(
    model="gpt-image-1",
    prompt="귀여운 고양이 귀 메이드가 미소를 지으며 손을 흔들고 있다",
    size="1536x1024",
    quality="high",
)

# 이미지 저장
image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)
with open("customize.png", "wb") as f:
    f.write(image_bytes)
```

# 참조 이미지를 사용한 이미지 편집

```python
import base64

# 이미지 편집
result = client.images.edit(
    model="gpt-image-1",
    image=open("cat.png", "rb"),
    prompt="귀여운 고양이가 블록 담 위에서 춤을 추고 있다"
)

# 이미지 저장
image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)
with open("cat-dance.png", "wb") as f:
    f.write(image_bytes)
```

# 마스크를 사용한 이미지 편집

```python
import base64

# 이미지 편집
result = client.images.edit(
    model="gpt-image-1",
    image=open("cat.png", "rb"),
    mask=open("mask.png", "rb"),
    prompt="골판지 상자 안에 귀여운 강아지가 있다",
)

# 이미지 저장
image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)
with open("dog.png", "wb") as f:
    f.write(image_bytes)
```
