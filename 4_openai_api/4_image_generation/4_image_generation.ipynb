```python
import os
from google.colab import userdata

# 環境変数の準備 (左端の鍵アイコンでOPENAI_API_KEYを設定)
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
```

```python
from openai import OpenAI

# クライアントの準備
client = OpenAI()
```

# テキストからの画像生成

```python
import base64

# 画像生成
result = client.images.generate(
    model="gpt-image-1",  # モデルID
    prompt="かわいい猫耳メイドが笑顔で手をふっている"  # プロンプト
)

# 画像の保存
image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)
with open("sample.png", "wb") as f:
    f.write(image_bytes)
```

# 画像出力のカスタマイズ

```python
import base64

# 画像の生成
result = client.images.generate(
    model="gpt-image-1",
    prompt="かわいい猫耳メイドが笑顔で手をふっている",
    size="1024x1024",
    quality="high",
)

# 画像の保存
image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)
with open("customize.png", "wb") as f:
    f.write(image_bytes)
```

# 参照画像を使った画像編集

```python
import base64

# 画像編集
result = client.images.edit(
    model="gpt-image-1",
    image=open("cat.png", "rb"),
    prompt="かわいいネコがブロック塀の上でダンスを踊ってる"
)

# 画像の保存
image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)
with open("cat-dance.png", "wb") as f:
    f.write(image_bytes)
```

# マスクを使った画像編集

```python
import base64

# 画像編集
result = client.images.edit(
    model="gpt-image-1",
    image=open("cat.png", "rb"),
    mask=open("mask.png", "rb"),
    prompt="ダンボールの中にかわいい子犬がいる",
)

# 画像の保存
image_base64 = result.data[0].b64_json
image_bytes = base64.b64decode(image_base64)
with open("dog.png", "wb") as f:
    f.write(image_bytes)
```
