```python
import os
from google.colab import userdata

# 環境変数の準備 (左端の鍵アイコンでOPENAI_API_KEYを設定)
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
```

```python
from openai import OpenAI

# クライアントの準備
client = OpenAI()
```

# 音声合成の実行

```python
# 音声合成の実行
response = client.audio.speech.create(
    model="gpt-4o-mini-tts",  # モデルID
    voice="alloy",  # 音声の種類
    input="ワガハイは猫である。名前はまだない。",  # 音声に変換するテキスト
    instructions="明るく前向きな口調で話しましょう。",  # 指示
)

# 音声ファイルに出力
response.stream_to_file("./speech.mp3")
```

```python
from IPython.display import Audio

# 音声ファイルの再生
Audio("speech.mp3")
```

# 指示による出力スタイルの変更

```python
# 音声合成の実行
response = client.audio.speech.create(
    model="gpt-4o-mini-tts",
    voice="alloy",
    input="ワガハイは猫である。名前はまだない。",
    instructions="悲しく後ろ向きな感じに話しましょう。",
)

# 音声ファイルに出力
response.stream_to_file("./speech2.mp3")
```

```python
from IPython.display import Audio

# 音声ファイルの再生
Audio("speech2.mp3")
```
