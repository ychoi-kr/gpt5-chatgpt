```python
import os
from google.colab import userdata

# 환경 변수 준비 (왼쪽 끝 키 아이콘으로 OPENAI_API_KEY 설정)
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
```

```python
from openai import OpenAI

# 클라이언트 준비
client = OpenAI()
```

# 음성 합성 실행

```python
# 음성 합성 실행
response = client.audio.speech.create(
    model="gpt-4o-mini-tts",  # 모델 ID
    voice="alloy",  # 음성 종류
    input="나는 고양이로소이다. 이름은 아직 없다.",  # 음성으로 변환할 텍스트
    instructions="밝고 긍정적인 어조로 말하세요.",  # 지시
)

# 음성 파일로 출력
response.stream_to_file("./speech.mp3")
```

```python
from IPython.display import Audio

# 음성 파일 재생
Audio("speech.mp3")
```

# 지시를 통한 출력 스타일 변경

```python
# 음성 합성 실행
response = client.audio.speech.create(
    model="gpt-4o-mini-tts",
    voice="alloy",
    input="나는 고양이로소이다. 이름은 아직 없다.",
    instructions="슬프고 부정적인 느낌으로 말하세요.",
)

# 음성 파일로 출력
response.stream_to_file("./speech2.mp3")
```

```python
from IPython.display import Audio

# 음성 파일 재생
Audio("speech2.mp3")
```
