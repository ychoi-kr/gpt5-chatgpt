```python
from datasets import load_dataset

# 데이터셋 준비
dataset = load_dataset("ychoikr/Ani-Bench-JP-ko", split="test")
print(dataset[0])
```

```python
import json

# JSONL로 변환
with open("ani-bench-jp-ko.jsonl", "w", encoding="utf-8") as f:
    for entry in dataset:
        question_text = f"「{entry['프로그램명']}」에 관한 문제입니다. {entry['문제']}"
        new_entry = {"question": question_text, "ground_truth": entry["답변"]}
        f.write(json.dumps(new_entry, ensure_ascii=False) + "\n")
```
