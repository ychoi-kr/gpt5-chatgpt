```python
import os
from google.colab import userdata

# 환경 변수 준비 (왼쪽 끝 키 아이콘으로 OPENAI_API_KEY 설정)
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
```

```python
from openai import OpenAI

# 클라이언트 준비
client = OpenAI()
```

# File search

```python
# 벡터 스토어 ID 준비 (자신의 벡터 스토어 ID 지정)
vector_store_id = "vs_..."
```

```python
# 메시지 리스트 준비
messages = [
    {
        "role": "user",
        "content": "GPT-5의 성능에 대해 알려줘.",
    }
]
```

```python
# 추론 실행
response = client.responses.create(
    model="gpt-5",
    tools=[
        {
            "type": "file_search",  # 툴
            "vector_store_ids": [vector_store_id]  # 벡터 스토어 ID
        }
    ],
    input=messages,
)
print(response.output_text)
```

# 인용 출처 확인

```python
# 인용 출처 확인
for item in response.output:
    # file_search_call
    if item.type == "file_search_call":
        print("==file_search_call==")
        print(item)

    # message
    if item.type == "message":
        print("==message==")
        print(item.content[0].text)
        print(item.content[0].annotations)
```
