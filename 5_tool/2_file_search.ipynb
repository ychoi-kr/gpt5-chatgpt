```python
import os
from google.colab import userdata

# 環境変数の準備 (左端の鍵アイコンでOPENAI_API_KEYを設定)
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
```

```python
from openai import OpenAI

# クライアントの準備
client = OpenAI()
```

# File search

```python
# ベクトルストアIDの準備 (自分のベクトルストアIDを指定)
vector_store_id = "vs_..."
```

```python
# メッセージリストの準備
messages = [
    {
        "role": "user",
        "content": "GPT-5の性能について教えて。",
    }
]
```

```python
# 推論の実行
response = client.responses.create(
    model="gpt-5",
    tools=[
        {
            "type": "file_search",  # ツール
            "vector_store_ids": [vector_store_id]  # ベクトルストアID
        }
    ],
    input=messages,
)
print(response.output_text)
```

# 引用元の確認

```python
# 引用元の確認
for item in response.output:
    # file_search_call
    if item.type == "file_search_call":
        print("==file_search_call==")
        print(item)

    # message
    if item.type == "message":
        print("==message==")
        print(item.content[0].text)
        print(item.content[0].annotations)
```
